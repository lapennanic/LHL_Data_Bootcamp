```python
import pandas as pd
import json
```

```python
with open('world_bank_projects.json') as f:
    bank_data = json.load(f)
```

```python
from pprint import pprint

pprint(bank_data)
```

```python
from pandas.io.json import json_normalize
json_normalize(bank_data)
```

```python
data = json_normalize(bank_data)
```

```python
data.info()
```

```python
# 1. Find the 10 countries with most projects
data['countryshortname'].value_counts().sort_values(ascending=False)[0:10]
```

```python
# 2. What are top 10 sectors with projects?
json_normalize(bank_data, record_path='sector').value_counts()[0:10]
```

```python
# 3. Find the top 10 major project themes (using column 'mjtheme_namecode')
json_normalize(bank_data, record_path='mjtheme_namecode').value_counts()[0:10]
```

```python
# 4. In 3. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

themes = json_normalize(bank_data, record_path='mjtheme_namecode')

# edit values back to the right types
unique_themes = themes[themes['name'] != ''].groupby(by=['code', 'name']).value_counts().reset_index().drop(columns=0)
unique_themes['code'] = pd.to_numeric(unique_themes['code'])
unique_themes['name'] = unique_themes['name'].astype('string')
unique_themes.set_index('code')
```

```python
unique_themes = unique_themes.sort_values(by='code')
unique_themes.to_dict()

themes_dict = unique_themes['name']
themes_dict
```

```python
themes_filled = themes['name'].fillna(value=themes_dict)

df_filled = pd.DataFrame(themes_filled)
df_filled[]
```
